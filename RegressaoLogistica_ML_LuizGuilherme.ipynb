```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('titanic_v2.csv', sep=',', decimal=',')
df
```

```python
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod =['Sex']

le = LabelEncoder()

for i in var_mod:
  df[i] = le.fit_transform(df[i])

df
```

```python
df_x = df.drop(axis=1, columns=['Survived', 'Name'])
df_x
```

```python
df_y = df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

nao tem over

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model = LogisticRegression().fit(X_train, y_train)
```

```python
predict = model.predict(X_train)

print(accuracy_score(y_train, predict))
print(confusion_matrix(y_train, predict))
```

**Acurácia: 82.10%**

A acurácia no conjunto de treinamento é bastante alta, o que pode sugerir que o modelo está se ajustando bem aos dados de treinamento.

```python
predict = model.predict(X_test)

print(accuracy_score(y_test, predict))
print(confusion_matrix(y_test, predict))
```

**Acurácia: 77.15%**

A acurácia no conjunto de teste é um pouco menor do que no conjunto de treinamento, o que é esperado. Isso sugere que o modelo pode estar generalizando bem, mas não tão bem quanto no treinamento.

**CONCLUSÃO**

**Desempenho do Modelo:**

*  **Treinamento**: O modelo tem uma alta acurácia de 82.10% no conjunto de treinamento. Isso indica que o modelo é bastante eficaz para classificar os dados com base nas características fornecidas.
*   **Teste**: A acurácia no conjunto de teste é de 77.15%, o que é um pouco menor, mas ainda é uma boa performance. A diferença entre a acurácia de treinamento e teste pode ser um indicativo de que o modelo não está excessivamente sobreajustado.

**Overfitting e Underfitting:**

* **Overfitting**: A diferença entre a acurácia do treinamento e do teste (4.95%) sugere uma leve tendência para overfitting. O modelo está performando melhor no conjunto de treinamento do que no conjunto de teste, o que pode indicar que ele está aprendendo muito bem as particularidades dos dados de treinamento, mas não generaliza perfeitamente para novos dados.
* **Underfitting**: Não há evidências claras de underfitting, pois a acurácia no conjunto de teste é relativamente boa. O modelo parece ser capaz de capturar padrões úteis nos dados.
