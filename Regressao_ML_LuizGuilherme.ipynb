```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('salarios_recife_outubro_2013.csv', sep=',', decimal=',')
df
```

```python
nulos_por_coluna = df.apply(lambda x: sum(x.isnull()), axis=0)
print(nulos_por_coluna)
```

```python
df['FUNCAO'].fillna(df['FUNCAO'].mode()[0], inplace=True)
```

```python
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
df['R$SALARIO'] = df['R$SALARIO'].str.replace('R$', '', regex=False).str.replace(',', '').astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod =['ORGAO','NOME','CATEGORIA','CARGO','FUNCAO','R$SALARIO']

le = LabelEncoder()

for i in var_mod:
  df[i] = le.fit_transform(df[i])

df
```

```python
df_x = df.drop(axis=1, columns=['NOME', 'R$SALARIO'])
df_x
```

```python
df_y = df[['R$SALARIO']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(df_x.columns)  # Verificando se ta td ok
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

model = LinearRegression().fit(X_train, y_train)
```

```python
y_predict = model.predict(X_train)

print(mean_absolute_error(y_train, y_predict))
print(mean_absolute_error(y_train, y_predict)/y_train.mean())
print(mean_squared_error(y_train, y_predict))
```

```python
y_predict = model.predict(X_test)

print(mean_absolute_error(y_test, y_predict))
print(mean_absolute_error(y_test, y_predict)/y_test.mean())
print(mean_squared_error(y_test, y_predict))
```

Os resultados indicam que o modelo está **generalizando de forma adequada**, com desempenho consistente tanto no conjunto de treinamento quanto no conjunto de teste. **A ausência de discrepâncias significativas** entre essas métricas sugere que o modelo está **equilibrado** e não apresenta sinais evidentes de **overfitting** ou **underfitting**.
