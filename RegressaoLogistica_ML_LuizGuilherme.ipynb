```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('titanic_v2.csv', sep=',', decimal=',')
df
```

```python
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod =['Sex']

le = LabelEncoder()

for i in var_mod:
  df[i] = le.fit_transform(df[i])

df
```

```python
df_x = df.drop(axis=1, columns=['Survived', 'Name'])
df_x
```

```python
df_y = df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

nao tem over

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model = LogisticRegression().fit(X_train, y_train)
```

```python
predict = model.predict(X_train)

print(accuracy_score(y_train, predict))
print(confusion_matrix(y_train, predict))
```

**AcurÃ¡cia: 82.10%**

A acurÃ¡cia no conjunto de treinamento Ã© bastante alta, o que pode sugerir que o modelo estÃ¡ se ajustando bem aos dados de treinamento.

```python
predict = model.predict(X_test)

print(accuracy_score(y_test, predict))
print(confusion_matrix(y_test, predict))
```

**AcurÃ¡cia: 77.15%**

A acurÃ¡cia no conjunto de teste Ã© um pouco menor do que no conjunto de treinamento, o que Ã© esperado. Isso sugere que o modelo pode estar generalizando bem, mas nÃ£o tÃ£o bem quanto no treinamento.
